## 1.10 이미지 resize 함수 하나 만들고

```python
def img_resize(img):
    img = color.rgb2gray(img[:,:,:3])
    return resize(img, (28,28))
```

## 1.11 train_O

```python
from tqdm.notebook import tqdm

def convert_train_O():
    train_raw_O_list =glob('./OX_classification/train_raw/O/*')
    for each in tqdm(train_raw_O_list):
        img = imread(each)
        img_resized = img_resize(img)
        save_name = './OX_classification/train/O/' + each.split('/')[-1][:-4] + '.png'
        imsave(save_name, np.round(img_resized*255).astype(np.uint8))

convert_train_O()
```

## 1.17 모듈 import

```python
import numpy as np
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.layers import Flatten
from tensorflow.python.keras.layers.convolutional import Conv2D
from tensorflow.python.keras.layers.convolutional import MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
import keras

np.random.seed(13)

keras.__version__
```

## 1.18 image_generator

```python
train_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    './OX_classification/train',
    target_size=(28,28),
    batch_size=3,
    class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    './OX_classification/test',
    target_size=(28,28),
    batch_size=3,
    class_mode='categorical')
```

## 1.19 모델 구성

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3),
                 activation='relu',
                 input_shape=(28,28,3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

## 1.20 summary

```python
model.summary()
```

## 1.21 compile

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam', metrics=['accuracy'])
```

## 1.22 fit

```python
hist = model.fit_generator(
    train_generator,
    steps_per_epoch=15,
    epochs=50,
    validation_data=test_generator,
    validation_steps=5)
```

## 1.23 학습 상황을 그려볼까

```python
plt.figure(figsize=(12,6))
plt.plot(hist.history['loss'], label='loss')
plt.plot(hist.history['val_loss'], label='val_loss')
plt.plot(hist.history['accuracy'], label='acc')
plt.plot(hist.history['val_accuracy'], label='val_acc')
plt.legend()
plt.show()
```

## 1.25 테스트 데이터에서

```python
scores = model.evaluate(
            test_generator,
            steps = 5)

print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

## 1.26 predict 확률도 엄청남

## 1.27

```python
n = 1

def show_prediction_result(n):
    img = imread(test_generator.filepaths[n])
    pred = model.predict(np.expand_dims(color.gray2rgb(img), axis=0))
    title = 'Predict : ' + str(np.argmax(pred))
    plt.imshow(img/255., cmap='gray')
    plt.title(title)
    plt.show()

show_prediction_result(n)
```
